```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
```

```
df=pd.read_csv('project.csv')
```

```
df
```

```
x = df.drop(['CLOSEP*'],axis=1)
x
```

```
y =df['CLOSEP*']
y
```

```
df.isnull().sum()
```

```
city = pd.get_dummies(x['TRADING_CODE'],drop_first=True)
city
```

```
x = x.drop('TRADING_CODE',axis=1)
x
```

```
x = pd.concat([x,city],axis=1)
x
```

```
x = x.replace('[^\d.]', '', regex=True).astype(float)
```

```
x.dtypes
```

```
y = y.replace('[^\d.]', '', regex=True).astype(float)
y.dtypes
```

```
from sklearn.model_selection import  train_test_split
```

```
xtrain, xtest, ytrain, ytest =train_test_split(x,y,test_size=0.25)
```

```
from sklearn.linear_model import LinearRegression
```

```
reg = LinearRegression()
```

```
reg.fit(xtrain,ytrain)
```

```
xtest
```

```
ytest
```

```
pred = reg.predict(xtest)
```

```
pred
```

```
reg.score(xtest,ytest)
```

```
compare = ytest
compare['prediction'] = pred
```
